# Featuring Engineering

Para a Regressão Linear vamos utilizar a area do lote e o preço de venda

```python
import pandas as pd
```

```python
df_limpo = pd.read_csv(r'..\train.csv')
df_limpo = df_limpo[['LotArea','SalePrice']]
df_limpo
```

# Codigo

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
from scipy.stats import pearsonr
```

```python
lot_area = np.array([df_limpo['LotArea'].values]).T
sale_price = df_limpo['SalePrice'].values
```

```python
regressao = LinearRegression()
regressao.fit(lot_area,sale_price)
```

```python
sale_price_est = regressao.predict(lot_area)
```

```python
plt.scatter(lot_area, sale_price, color='blue')
plt.plot(lot_area,sale_price_est, color= 'black', linewidth=2)
plt.xlabel('Area do Lote')
plt.ylabel('Preço de Venda')
plt.title('Regressão: Area do Lote x Preço De Venda')
plt.show()
```

# Metricas de Avaliação

```python
fitted = regressao.predict(lot_area)
RMSE = np.sqrt(mean_squared_error(df_limpo['SalePrice'].values, fitted))

r2 = r2_score(df_limpo['SalePrice'].values, fitted)

mae = mean_absolute_error(df_limpo['SalePrice'].values, fitted)

print(f"RMSE: {RMSE}")
print(f"R2: {r2}")
print(f"MAE: {mae}")
```
